## Transformações Lineares

**Definição:** Sejam dois espaços vetoriais $V$ e $W$, uma transformação linear é uma função de $V$ em $W$, $F: V \rightarrow W$ que satisfaz as seguintes condições:
1. Para quaisquer ${\bf u}\ e\ {\bf v} \in V: F({\bf u} + {\bf v}) = F({\bf u}) + F({\bf v})$
2. Para qualquer $k \in \mathbb{R}$ e ${\bf v} \in V:$ $F(k{\bf v}) = kF({\bf v})$

**Exemplo:** Sejam $F: \mathbb{R} \rightarrow \mathbb{R}$ definida por $F(x, y, z) = (x, 2x - z),\ \forall {\bf v} = (x, y, z) \in \mathbb{R}^3$, temos que dados ${\bf w} = (x_1, y_1, z_1)$ e ${\bf z} = (x_2, y_2, z_2) \in \mathbb{R}^3$:

1. $$F[{\bf w} + {\bf z}] = F(x_1 + x_2, y_1 + y_2, z_1 + z_2) = (x_1 + x_2, 2(x_1 + x_2) - (z_1 + z_2)) = (x_1, 2x_1 - z_1) + (x_2, 2x_2 - z_2) = F({\bf w}) + F({\bf z})$$ 

2. $F(k{\bf w}) = F(k x_1, k y_1, k z_1) = k( x_1, 2 k x_1 - k z_1) = k(x_1, 2x_1 - z_1)$

Logo é uma transformação linear.

**Exemplo:** Seja $F: \mathbb{R} \rightarrow \mathbb{R}^2$ sendo $F(x) = (x, 2),\ \forall x \in \mathbb{R}$, teremos:

1. $F(x + y ) = (x + y, 2) \neq (x,2) + (y, 2) = (x+y, 4)$

logo não é uma transformação linear.

**Exemplo:** Sabendo que  $F: \mathbb{R}^2 \rightarrow \mathbb{R}^2$ é um operador linear e que $F(1, 2) = (3, -1)$ e $F(0, 1) = (1,2)$, determinar $F(x, y)$ sendo $(x, y) \in \mathbb{R}^2$.

Inicialmente temos que $\{ (1,2), (0,1)\}$ e uma base de $\mathbb{R}^2$, logo vamos determinar as coordenadas de $(x, y) \in \mathbb{R}^2$ nesta base:

$$(x, y) = a(1,2) + b(0,1)= (a, 2a+ b) \rightarrow a = x,\ b = y - 2x$$

Logo: $(x, y)  = x(1, 2) +(y - 2x)(0,1)$, portanto

$$F(x, y) = xF(1, 2) + (y - 2x) f(0, 1) =  x(3, -1) + (y - 2x)(1, 2)$$

$$F(x, y) = (x+y, -5x + 2y)$$

**Exemplo:** Sejq $P$ uma matriz inversível de $M_n(\mathbb{R})$, ou $M_n\times n$. Mostrar que $F:M_{n\times n} \rightarrow M_{n\times n}$ dada por $F(X) = P ^{-1}XP$ é um operador linear.

1. $F(X + Y) = P^{-1}(X + Y)P = P^{-1}XP + P^{-1}XP = F(X) + F(Y)$
2. $ F(\alpha X) = P^{-1}(\alpha X)P = \alpha(P^{-1}XP) = \alpha F(X)$

Observação: Se $T \rightarrow W$ é uma transformação linear, se $0 \in W$, então $T(0) = 0$, portanto se $T(0) \neq 0$ a transformação não é linear. Porém $T(0) = 0$ não garante que $T$ seja linear, por exemplo $\mathbb{R} \rightarrow \mathbb{R}$ tal que $T({\bf u}) = {\bf u} ^2$, que tem $T(0) = 0$ mas

$$T({\bf u} + {\bf v}) = ({\bf u} + {\bf v})^2 = {\bf u}^2 + 2{\bf u}{\bf v} + {\bf v}^2 \neq T{\bf u} + T{\bf v}$$
logo não é linear.

**Teorema:** Dados dois espaços vetoriais $V$ e $W$, uma base de $V$, $\{{\bf v}_1, ..., {\bf v}_m\}$ e ${\bf w}_1, {\bf w}_2, ...,  {\bf w}_n$ elementos arbitrários de $W$. Então existe uma única transformação linear $T:V \rightarrow W$ tal que $T({\bf v}_1) = {\bf w}_1$, ..., $T({\bf v}_n) = {\bf w}_n$, dada por

$$T({\bf v}) = a_1T({\bf v}_1) + ... + a_n T({\bf v}_n) = a_1{\bf w}_1 + ... + a_n {\bf w}_n$$

**Exemplo** Seja $T: \mathbb{R}^2 \rightarrow \mathbb{R}^3$ tal que $T(1,0) = (2, -1, 0)$ e $T(0, 1) = (0, 0, 1)$. Qual é a transformação linear?

Neste caso ${\bf e}_1 = (1,0)$ e ${\bf e}_2 = (0,1)$ formam uma base para o $\mathbb{R}^2$, logo
${\bf v} \in \mathbb{R}^2 \rightarrow {\bf v} = x_1 {\bf e}_1 + x_2{\bf e}_2$
e 
$$T({\bf v}) = x_1T({\bf e}_1) + x_2 T({\bf e}_2) = x_1(2, -1, 0) + x_2(0, 0, 1)$$
$$T({\bf v}) = (2 x_1 - x_2, x_2)$$

## Transformações Lineares do Plano no Plano

 
**Expansão (ou contração) uniforme:** $T: \mathbb{R}^2 \rightarrow \mathbb{R}^2$, $\alpha \in \mathbb{R}$, ${\bf v} \rightarrow \alpha {\bf v}$. Exemplo: ${\bf v} \rightarrow 2{\bf v}$, ou $T(x, y) = 2(x, y)$, ou $T_1({\bf v}) = \frac{1}{2}(x, y)$, e essa transformação pode ser escrita como

$$
\left[
\begin{array}{c} x \\ y \end{array} \right]
\rightarrow \frac{1}{2}\left[\begin{array}{c} x \\ y \end{array} \right]
$$
ou 
$$
\left[
\begin{array}{c} x \\ y \end{array} \right]
\rightarrow \left[\begin{array}{c c} \frac{1}{2} & 0 \\ 0 & \frac{1}{2} \end{array}\right]
\left[\begin{array}{c} x \\ y \end{array} \right]
$$

**Reflexão em torno do eixo:** $F: \mathbb{R}^2 \rightarrow \mathbb{R}^2 : (x, y) \rightarrow (x, -y)$
$$
\left[
\begin{array}{c} x \\ y \end{array} \right]
\rightarrow \left[\begin{array}{c} x \\ -y \end{array} \right]
$$
ou 
$$
\left[
\begin{array}{c} x \\ y \end{array} \right]
\rightarrow \left[\begin{array}{c c} 1 & 0 \\ 0 & -1 \end{array}\right]
\left[\begin{array}{c} x \\ y \end{array} \right]
$$

**Rotação de um ângulo $\theta$ (sentido anti-horário):**
$$
\left[
\begin{array}{c} x \\ y \end{array} \right]
\rightarrow \left[\begin{array}{c c} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{array}\right]
\left[\begin{array}{c} x \\ y \end{array} \right]
$$

Se $\theta = \pi/2$
$
\left[
\begin{array}{c} x \\ y \end{array} \right]
\rightarrow \left[\begin{array}{c c} 0 & -1 \\ 1 & 0 \end{array}\right]
\left[\begin{array}{c} x \\ y \end{array} \right]
$

### Núcleo e Imagem

**Definição:** Seja $T: V \rightarrow W$ uma transformação linear. A imagem de $T$ é o conjunto de vetores $w \in W$ tais que existe um vetor $v \in V$ que satisfaz $T(v) = w$. Formalmente
$$
Im(T) = \{ w \in W; T(v) = w, para\ algum\ v \in W\}
$$

**Definição:** O núcleo (Kernel) de uma transformação linear $T: V \rightarrow W$ é definido como
$$
Ker(T) = \{ {\bf v} \in V, T({\bf v}) = 0\}
$$

**Exemplo** Seja $F: {\mathbb R}^2 \rightarrow {\mathbb R}^3$ dada por $F(x, y) = (0, x+y, 0)$ então o núcelo de $F$ é dado por
$$
(0, x+y, 0) = (0, 0, 0) \rightarrow x=-y
$$
logo $Ker(F) =  \{(x, -x) | x \in {\mathbb R}\}$.

**Exemplo** O operador linear $D: P_n({\mathbb R} \rightarrow P_n({\mathbb R})$ dado $D[f(t)] = f^\prime(t)$, então temos
$$
f(t) = a_0 + a_1t + ... + a_nt^n
$$
logo $D[f(t)] = a_1 + 2a_2t + ... _ na_nt^{n-1}$, e para $f(t) = 0$ devemos ter $a_1 = a_2 = ... =a_n = 0$, e portanto $ker(D) = \{a_0 |\ a_0 \in {\mathbb R}\}$. 

**Exemplo** Seja $F: {\mathbb R}^2 \rightarrow {\mathbb R}^3$, definido por $F(x, y) = (x, 0)$, temos que 
1. $kerf(F) = \{ (x, y) \in \mathbb{R}^2 | F(x, y) = (0,0)\} = \{(0, y) | y \in {\mathbb R}\}$
2. $Im(F) = \{(x, 0) | x \in {\mathbb R}\}$. Logo $Im(F)$ é o eixo $x$.

Observações:
1. A imagem de uma transformação é um subconjunto de $W$, e é um subespaço vetorial de $W$.
2. O núcleo de uma transformação é um subconjunto de $V$, e é um subespaço vetorial de $V$.

### Transformações Injetoras e Sobrejetoras

Relembrando:
1.  Uma função $f: V \rightarrow W$ é injetora se $\forall {\bf u} \in V$ e ${\bf v} \in V$, ${\bf u} \neq {\bf v}$ então $f({\bf u}) \neq f({\bf v})$.
2. Uma função será sobrejetora se a imagem de $f$ coincide com $W$, ou seja, $T(V) = W$.

**Definição:** Uma aplicação linear $T:V \rightarrow W$ é injetora se, dados ${\bf u} \in V$, ${\bf v} \in V$, com $T({\bf u}) = T({\bf v})$ então ${\bf u} = {\bf v}$.
\end{definicao}

**Definição:** Uma transformação linear $T: V \rightarrow W$ será sobrejetora se a imagem de $T$ coincidir com $W$, ou seja, $T(V) = W$.



Quando uma T.L. $T: V \rightarrow W$ for injetora e sobrejetora, ao mesmo tempo, tem-se um isomorfismo, e dizemos que esses espaços são isomorfos, o que significa que são idênticos. Um isomorfismo leva {\bf base em base}. Além disso, um isomorfismo tem uma aplicação inversa $T^{-1} : W\rightarrow V$.

**Exemplo** Seja a transformação $F: \mathbb{R}^3 \rightarrow \mathbb{R}^3$ dada por $F(x, y, z) = (x + y, x-z, y)$. Mostre que $F$ é um isomorfismo e determine $F^{-1}$.


Vamos achar o núcleo de $F$: $F(x, y, z) = (0, 0, 0) \rightarrow (x+y, x-z, y) = (0, 0, 0)$.
$$
\left\{
\begin{array}{c} 
x+ y =0 \\
x -z = 0 \\
y = 0 \\
\end{array}
\right.
$$
cuja única solução é $(0, 0, 0)$. Como o $Ker(F) = \{(0, 0, 0)\}$ $F$ é injetora, sendo $Dim(V) = Dim(W)$, $F$ é sobrejetora, logo é um isomorfismo.

Supondo que $F^{-1}(x, y, z) = (a, b, c)$, então $(x, y, z) = F(a, b, c) = (a+c, a-c, b)$, logo
$$
\left\{
\begin{array}{c}
a + c = x \\
a - c = y \\
b = z
\end{array}
\right.
$$
$a = \frac{x+y}{2}$, $b=z$ e $c = \frac{x-1}{2}$, e $F^{-1}(x, y, z) = \left(\frac{x+y}{2}, z, \frac{x-y}{2}\right)$.

**Teorema:** Seja $T:\ V\rightarrow W$ uma aplicação linear. Então $Ker(T) = \{0\}$ se e somente se $T$ é injetora.


**Prova**
1. Se $Ket(T) = \{0\}$ e $\bf u$ e $\bf v$ $\in V$, como $T({\bf u}) = T({\bf v}) \rightarrow T({\bf u}) - T({\bf v}) =) \rightarrow T({\bf u} - {\bf v}) =0$. Logo ${\bf u} - {\bf v} \in Kert(T)$, porém como o único elemento do núcleo é $0$ temos ${\bf u} - {\bf v} = 0 \rightarrow {\bf u} = {\bf v}$, $T$ é injetora.
2. Assumindo que $T$ é injetora, então $ker(T) = \{0\}$. Então seja ${\bf v}\in Kert(T)$, ou seja $T({\bf v}) = 0$. Como necessariamente $T(0) = 0 \rightarrow T(v) = T(0)$, logo ${\bf v} = 0$, logo ${\bf v} = 0$, pois $T$ é injetora, portanto $ker(T) = \{0\}$.

**Obs.:** Uma aplicação linear injetora leva vetores LI em vetores LI.

Quando um T.L. $T: V \rightarrow W$ for injetora e sobrejetora, ao mesmo tempo, tem-se um isomorfismo, e dizemos que esses espaços são isomórfos, o que significa que são idênticos. Um isomorfismo leva {\bf base em base}. Além disso, um isomorfismo tem uma aplicação inversa $T^{-1}$

**Exemplo** Seja a transformação $F: {\mathbb R}^3 \rightarrow {\mathbb R}^3$ dada por $F(x, y, z) = (x+y, x-z, y$. Mostre que $F$ é um isomorfismo e determine $F^{-1}$.


**Solução:** Vamos determinar o núcleo de $F$: 
$F(x, y, z) = (0, 0, 0) \rightarrow *x+y, x-z, y) = (0, 0, 0)$ de onde se obtém
$$
\left\{ \begin{array}{c} x + y =0 \\ x - z = 0 \\ y =0 \end{array} \right.
$$
cuja única solução é $(0, 0, 0)$, e como $ker(F) = \{(0, 0, 0)\}$, $F$ é injetora, e sendo $Dim(V) = Dim(W)$, $F$ é sobrejetora, logo é um {\bf isomorfismo}.

Supondo que $F^{-1}(x, y, z) = (a, b, c)$, então $(x, y, z) = F(a, b, c) = (a+c, a-c, b)$ logo
$$
\left\{ \begin{array}{c} a+ c =x \\ a - c = y \\ b = z \end{array} \right.
$$
com solução $a = (x+y)/2$, $b = z$ e $c = (x-1)/2$, logo
$$
F^{-1}(x, y, z) = \left(\frac{x+y}{2}, z, \frac{x-y}{2} \right)
$$
 
**Exemplo** $F: {\mathbb R}^3 \rightarrow {\mathbb R}^4$ dada por $F(x, y, z) = (x, x-1, y-z, z)$ é um isomorfismo?

**Solução:** Deve-se verificar se $F$ é uma transformação linear. O núcleo de $F$ é dado por
$$
\left\{ \begin{array}{c} x = 0 \\ x - y = 0 \\ y - z = 0 \\ z = 0 \end{array} \right.
$$
Logo $F$ é injetora. Mas $F$ não é sobrejetora, pois $Dim[Im(F)] = Dim({\mathbb R}^3) - Dim[Ker(F)] = 3 - 0 = 3 \neq Dim({\mathbb R}^4)$. Logo $F$ não é um isomorfismo.
