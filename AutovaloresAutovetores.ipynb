# <font color ="0000ff">Autovetores e Autovalores

**Definicao:**  Seja $T: V \rightarrow V$ um operador linear, se existirem ${\bf v}\in V$, ${\bf v} \neq 0$ e $\lambda \in \mathbb{R}$ tais que $T({\bf v}) = \lambda {\bf v}$, $\lambda$, é denominado de autovalor de $T$, e $\bf v$ um autovetor associado à $\lambda$.


Obs.: Visão geométrica, $v$ e $\lambda v$ tem a mesma direção.

**Exemplo:** Seja $T: \mathbb{R}^2 \rightarrow \mathbb{R}^2$ definida por $T({\bf v}) = 2{\bf v}$. Essa transformação pode ser representada por 
$
\left[
\begin{array}{c} x \\ y \end{array} \right] \rightarrow \left[ \begin{array}{c c} 2 & 0 \\ 0 & 2 \end{array}\right]\left[\begin{array}{c}x \\ y\end{array}\right] = \left[\begin{array}{c}2x \\ 2y\end{array}\right] = 2\left[\begin{array}{c}x \\ y\end{array}\right]
$
logo 2 é um autovalor e qualquer $(x, y) \neq (0,0)$ é um autovetor associado à $\lambda = 2$.


**Exemplo:** $F: \mathbb{R}^2 \rightarrow \mathbb{R}^2$, sendo $F(x, y) = (-y, x)$. Podemos escrever
$
\left[\begin{array}{c}x \\ y\end{array}\right] \rightarrow \left[\begin{array}{c c} 0 & -1\\ 1 & 0 \end{array}\right]\left[\begin{array}{c}x \\ y\end{array}\right] = 1\left[\begin{array}{c}-y \\ x\end{array}\right]
$
e pode-se notar que é impossível que $F(x, y) = \lambda(x, y)$, logo $F$ não tem autovalores.


**Teorema:** Dada uma transformação $T:V \rightarrow V$ e um autovetor $\bf v$ associado a um autovalores $\lambda$, qualquer vetor ${\bf w} = \alpha{\bf v}$ ($\alpha \neq 0$) também é um autovetor associada à $\lambda$.


**Ddefinicao:**  O subespaço vetorial $V_\lambda = \{{\bf v} \in V: T({\bf v}) = \lambda{\bf v}\}$ é denominado de subespaço associado ao autovalor de $\lambda$.


Dada uma matriz quadrada $A$, de ordem $n$, associaremos essa matriz à transformação $T: \mathbb{R}^n \rightarrow \mathbb{R}^n$ em relação à base canônica, ou se $T_A({\bf v}) = A{\bf v})$,  e um autovalor $\lambda \in \mathbb{R}$ de $A$ e um autovetor ${\bf v} \in \mathbb{R}^n$, são soluções da equação $A{\bf v} = \lambda{\bf v}$, ${\bf v} \neq 0$.

Vamos denominar simplesmente de autovalores e autovetores da matriz $A$.

**Exemplo:** Determine os autovalores e os respectivos autovetores da matriz
$
A = \left[
\begin{array}{c c} 1 & 2 \\ 3 & 2 \end{array}\right]
$

Vamos fazer $A{\bf v} = \lambda{\bf v}$ com ${\bf v} = \left[\begin{array}{c}x \\ y\end{array}\right]$
$
\left[\begin{array}{c c} 1 & 2 \\ 3 & 2\end{array}\right]\left[\begin{array}{c} x \\ y\end{array}\right] = \lambda\left[\begin{array}{c} x \\ y\end{array}\right] \rightarrow
\left[\begin{array}{c} x + 2y \\ 3x+2y\end{array}\right] = \left[\begin{array}{c} \lambda x \\ \lambda y\end{array}\right]
$

$
\left\{\begin{array}{c} x + 2y = \lambda x \\ 3x + +2y = \lambda y \end{array}\right.
\rightarrow
\left\{\begin{array}{c} (\lambda - 1)x - 2y = 0 \\ -3x + (\lambda - 2)y = 0 \end{array}\right.
$

Um sistema homogêneo tem solução não nula, se e somente se, o determinante da matriz de coeficientes é zero, logo
$
\left|\begin{array}{c c}\lambda -1 & -2 \\-3 & \lambda-2\end{array}\right| = \lambda^2 - 3\lambda - 4 = (\lambda -4)(\lambda + 1) = 0
$
Logo os autovalores são $\lambda_1 = 4$ e $\lambda_2 = -1$.

Fazendo $\lambda = 4$ temos
$
\left\{\begin{array}{c} x + 2y = \lambda x \\ 3x + +2y = \lambda y \end{array}\right. \rightarrow 3x -2y = 0.
$
logo $v_1 = \left[\begin{array}{c}2 \\ 3\end{array}\right]$

E para $\lambda = -1$, temos $v_2 =  \left[\begin{array}{c}1 \\ -1\end{array}\right]$

## <font color ="0000ff">Polinômio Característico</font>

**Definicao:** Dada uma matriz $A$ quadrada $n\times n$, a equação $det(\lambda I - A) = 0$ é denominada equação característica de $A$, e o polinômio $P(\lambda) = det(\lambda I -A$) de polinômio característico.


As raízes do polinômio característico são os autovalores da matriz $A$.

**Exemplo** Seja $A = \left[ \begin{array}{c c c} 1 & -3 & 3 \\ 3 & -5 & 3 \\6 & -6 &4\end{array}\right]$, escrevendo o polinômio característico temos
$
p(\lambda) = det \left\{ \left[ \begin{array}{c c c} \lambda & 0 & 0 \\ 0 & \lambda & 0 \\ 0 & 0 & \lambda \end{array} \right] - \left[ \begin{array}{c c c} 1 & -3 & 3 \\ 3 & -5 & 3 \\6 & -6 &4\end{array}\right]\right\} =  det\left\{ \left[ \begin{array}{c c c} \lambda -1 & 3 & -3 \\ -3 & \lambda + 5 & -3 \\ -6 & 6 & \lambda - 4 \end{array} \right]\right\}
$
$
= (\lambda + 2)^2(\lambda - 4)
$
cujas raízes são $\lambda_1 = -2$ e $\lambda_2 = 4$.

Fazendo $\lambda = -2$ obtemos
$
\left\{ \begin{array}{c} -3x + 3y -3z =  0\\-3x + 3y -3z =0 \\-6x + 6y - 6z = 0\end{array}\right.
$
e o sistema tem duas variáveis independentes, logo podemos escrever $x = y + z$, e os vetores ${\bf v} = \left[\begin{array}{c} y+z \\ y \\ z \end{array} \right]$ são autovetores associados à $\lambda = -2$.


Fazendo $\lambda = 4$ obtemos
$
\left\{ \begin{array}{c} 3x + 3y -3z =  0 \\ -3x + 9y - 3z =0 \\ -6x + 6y  = 0\end{array}\right.
$
ou seja, $x+ y -z =0$ e $2y - z = 0$. Logo o sistema tem uma vairável livre, e podemos  escrever $y = z/2$, $x = z/2$ e os autovetores associados à $\lambda = 4$ são ${\bf v} = \left[\begin{array}{c} x \\ x \\ 2x \end{array} \right]$.


Observe que para $\lambda = 2$, temos que os autovetores são da forma ${\bf v} = \left[\begin{array}{c} y+z \\ y \\ z \end{array} \right]$ , e se escolhermos
* $y = 1$, $z = 0$ temos ${\bf v}_1= \left[\begin{array}{c} 1\\ 1 \\ 0 \end{array} \right]$ 
* $y = 0$, $z = -1$ temos ${\bf v}_2 = \left[\begin{array}{c} 1 \\ 0 \\ -1 \end{array} \right]$ 
e $\{ {\bf v}_1, {\bf v}_2\}$ formar uma base para o subespaço de autovetores associados à $\lambda = -2$.

Para $\lambda = 4$ temos ${\bf v} = \left[\begin{array}{c} x \\ x \\ 2x \end{array} \right]$ logo
$x = 1$ ${\bf v} = \left[\begin{array}{c} 1\\ 1 \\ 2 \end{array} \right]$  e $\{[1, 1, 2]\}$ é uma base para o subespaço de autovetores associado $\lambda = 4$.

Dependendo da ordem da matriz, é muito trabalhos calcular os autovalores e os respectivos autovetores. Para tanto, podemos fazer uso de software que contém funções que realizam esse cálculo. Um desses software é o Scilab (ww.scilab.org), no qual a finção \verb"spec" quando aplicada a uma matriz $A$, \verb"[avl, avt] = spec(A)"
retorna os autovalroes e os autovetores da matriz, guardado respectivamente em \verb"avl"e \verb"avt".

Por exemplo, dada a matriz $A = \left[\begin{array}{c c}1 & -1 \\ -1 &1\end{array}\right]$, obtemos
$
avl = \left[\begin{array}{c c}0 & 0 \\0 & 2\end{array}\right]
$
e
$
avt =\left[\begin{array}{c c}-0.707 & -0.707 \\-0,707 & 0.707\end{array}\right]
$

## <font color ="0000ff">Aplicação:</font>


Seja $A_{m\times n}$, $\lambda$ um autovalor de $A$ e $\bf x$ um autovetor associado à $\lambda$. Considere a sequência de multiplicações:
$$\begin{matrix}
A{\bf x} &=& \lambda{\bf x} \nonumber \\
A^2{\bf x} &=& A(A{\bf x}) = A\lambda {\bf x} = \lambda A{\bf x} = \lambda^2{\bf x} \nonumber  \\
\vdots \nonumber \\
A^k{\bf x} &=& A(\lambda^{k-1}{\bf x}) = ... = \lambda^k{\bf x} \nonumber
\end{matrix}
$$

**Exemplo:** Determine os autovalores e autovetores da matriz $A = \left[\begin{array}{c c} 7 & 4 \\ 3 & 6 \end{array} \right]$.

Vamos resolver a equação característica
$$\begin{eqnarray}
det(\lambda I - A) &=& 0 \nonumber \\
det\left[\begin{array}{c c}\lambda-7 & -4 \\-3 & \lambda-6 \end{array}\right] = 0 \nonumber \\
(\lambda-7)(\lambda-6) - (-3)(-4) &=& 0 \nonumber \\
\lambda^2-13\lambda+30 &=& 0 \nonumber \\
(\lambda-3)(\lambda-10) &=& 0 \nonumber
\end{eqnarray}
$$
Para $\lambda_1 = 3$ temos
$
A - 3I = \left[\begin{array}{c c}7-3 & 4 \\ 3 & 6 - 3 \end{array}\right] = \left[\begin{array}{c c} 4 & 4 \\ 3 & 3 \end{array}\right] =\left[\begin{array}{c c} 1 & 1 \\ 0 & 0 \end{array}\right]
$
$
\left[\begin{array}{c c} 1 & 1 \\ 0 & 0 \end{array}\right] = \left[\begin{array}{c} 0 \\ 0 \end{array}\right]
$
logo $x_1 = x_2$, ou seja $x_2 = -x_1$. Então $\left[\begin{array}{c} x_1 \\ -x_1 \end{array}\right]$ ou $x_1\left[\begin{array}{c} 1 \\ 1 \end{array}\right]$ indica que $\left[\begin{array}{c} 1 \\ 1 \end{array}\right]$ é o autovetor associado à $\lambda_1 = 3$. Veja que todos os vetores do conjunto $\{(1,1)\}$ são autovetores associados à $\lambda_1 = 3$.

Para $\lambda_2 = 10$: 
$
A - 10I = \left[\begin{array}{c c}7-10 & 4 \\ 3 & 6 - 10 \end{array}\right] = \left[\begin{array}{c c} -3 & 4 \\ 3 & -4 \end{array}\right] =\left[\begin{array}{c c} 1 & -4/3 \\ 0 & 0 \end{array}\right]
$
logo
$
\left[\begin{array}{c c} 1 & -4/3 \\0 & 0 \end{array}\right]\left[\begin{array}{c}x_1 \\ x_2 \end{array}\right] = \left[\begin{array}{c} 0 \\ 0 \end{array}\right]
$
$x_1  -\frac{4}{3}x_2 = 0\rightarrow x_2 = \frac{3}{4}x_1 \rightarrow x_1\left[\begin{array}{c}\end{array}\right]$ e $(1, 3/4)$ é um autovetor associado à $\lambda_2 = 10$.

**Exemplo:** Determine os autovalores da matriz
$
\left[
\begin{array}{c c c} 2 & 1 & 1 \\0 & 1 & 1 \\ 0 & 0 & 1 \end{array}
\right]
$

**Solução:** Como $A$ é uma matriz diagonal, o seu determinante é a multiplicação dos elementos da diagonal principa. A matriz $\lambda I - A$ também será diagonal, e os elementos da diagonal principal serão $\lambda - a_{ii}$, logo para a matriz acima temos
$
(\lambda - 2)(\lambda -1)(\lambda +1) = 0 \rightarrow \left\{\begin{array}{c} \lambda_1 = 2 \\ \lambda_2 = 1\\ \lambda_3 = 1 \end{array}\right.
$

**Teorema:**
Se $B = kA + dI$, sendo $k$ e $c \neq 0$, então os autovalores de $B$ são da forma $\mu = k\lambda + d$, sendo $\lambda$ os autovalroes de $A$, e os autovetores de $A$ e $B$ são idênticos.


**Exemplo:** Seja $A = \left[\begin{array}{c c}7 & 4 \\3 & 6\end{array}\right]$, com os autovalores $\lambda_1 = 3$ e $\lambda_2 = 10$, e os autovetores ${\bf x}_1 = (-1, 1)$ e ${\bf x}_2 = (4, 3)$. Para $B = 3A + 4I$ os autovalores são $\mu_1 = 3.3+4 = 13$ e $\mu_2 = 3.10 + 4 = 34$, com autovetores correspondentes $y_1 = (-1, 1)$ e $y_2 = (4, 3)$.

$
R = \left[\begin{array}{c c} \cos\theta & -\sin\theta \\ \sin\theta & \cos\theta \end{array}\right].
$
Determine os autovalores e autovetores para cada $\theta$.

**Solução:** O polinômio característico será 
$
p(\lambda) = det\left[\begin{array}{c c} \lambda - \cos\theta & \sin\theta \\ -\sin\theta & \lambda - \cos\theta \end{array}\right] = (\lambda - \cos\theta)^2 + \sin^2\theta = \lambda^2 - 2\lambda\cos\theta + \cos^2\theta + \sin\theta = \lambda^2 -2\lambda\cos\theta + 1
$
$\Delta = 4\cos^2\theta -4 = 4(cos^2\theta -1) = -4\sin\theta$, e $\Delta \leq 0$ (raízes complexas) a não ser que $\theta =0$ ou \theta = \pi$ o que implica em $\Delta = 0$, então
* Se $\theta = 0$ então $R = \left[\begin{array}{c c} 1 & 0 \\ 0 & 1 \right] = I_{2 \times 2}$, cujo único autovalor é 1, qual

**Teorema:** Uma matriz quadrada $A$ é inversível se, e somente se, $\lambda = 0$ não é um autovalor de $A$.

**Prova:** Para uma matriz $n\times n$ a equação característica
$
$\lambda^n + c_1\lambda^{n-1} + ... + c_n = 0
$
tem $\lambda = 0$ como solução se, e somente se, $c_n=0$. Mas
$
det(\lambda I-A) = \lambda^n + c_1\lambda^{n-1} + ... + c_n
$
logo, para $\lambda = 0$ temos $det(-A) = c_n$, ou $(-1)^n det(A) = c_n$. Assim, se $c_n = 0$, $det(A) =0$, e a matriz não tem inversa.

## <font color ="0000ff"> Diagonalização de Operadores </font>

A forma mais simples de representar um operador $[T]^\beta_\beta$ é por uma matriz diagonal. Para tanto, vamos buscar uma base na qual a matriz de transformação seja diagonal.

**Teorema:** Autovetores associados a autovalores distintos são linearmente independentes.

**Prova:** Sejam $\lambda_1$, $\lambda_2$, ..., $\lambda_n$ os $n$ autovalores distintos e ${\bf v}_1$, ${\bf v}_2$, ..., ${\bf v}_n$ os autovetores associados, formando a equação
\[
a_1{\bf v}_1 + a_2{\bf v}_2 + ... + a_n{\bf v}_n = {\bf 0}
\]
aplicando a transformação $T - \lambda_i{\bf I}$ temos:
$
\begin{eqnarray}
(T - \lambda_i{\bf I})(a_1{\bf v}_1 + a_2{\bf v}_2 + ... + a_n{\bf v}_n) = {\bf 0} \nonumber \\
a_1(T[v_1] - \lambda_i {\bf I}v_1) + a_2(T[v_2] - \lambda_i {\bf I}v_2) + ... + a_n(T[v_n] - \lambda_i {\bf I}v_n) = {\bf 0} \nonumber \\
a_1(\lambda_1{\bf v}_1 - \lambda_i {\bf v}_1) + a_2(\lambda_2{\bf v}_2 - \lambda_i {\bf v}_2) + ... + a_n(\lambda_n {\bf v}_n - \lambda_i{\bf v}_n) = {\bf 0} \nonumber \\
a_1(\lambda_1 - \lambda_i ){\bf v}_1 + a_2(\lambda_2 - \lambda_i){\bf v}_2 + ... + a_n(\lambda_n- \lambda_i){\bf v}_n = {\bf 0} \nonumber \\
\end{eqnarray}
$
Como $\lambda_1 \neq \lambda_i$, $\lambda_1 - \lambda_1 \neq 0$, e ${\bf v}_i \neq {\bf 0}$ tem-se que $a_i = 0$, logo os vetores são LI.


**Corolario:** Se $V$ é um espaço vetorial de dimensão $n$ e $T: V \rightarrow V$ é um operador linear que possui $n$ autovalores distintos, então $V$ possui uma base formada pelos autovetores de $T$.


**Exemplo:** Seja $T: {\mathbb R}^2 \rightarrow {\mathbb R}^2$ definida por $t(x, y)  = (-3x + 4y, -x + 2y)$. A matriz dessa transformação em relação à base canônica é
$
\begin{eqnarray}
T(1, 0) &=& (-3, -1) = -3(1,0) + (-1)(0,1) \nonumber \\
T(0, 1) = (4, 2) = 4(1,0) + 2(0,1) \nonumber
\end{eqnarray}
$
logo $[T]_\alpha^\alpha = \left[\begin{array}{c c} -3 & 4 \\ -1 & 2\end{array}\right]$.
e para determinar os autovalores $det(\lambda I - A) = det\left|\begin{array}{c c} \lambda + 3 & -4 \\1 & \lambda -2\end{array}\right] = (\lambda + 3)(\lambda -2) = 0$ e $\lambda_1 = 1$ com autovetores $x(1,1)$, $\lambda_2 = -2$ com autovetores $y(4, 1)$.

Podemos observar que os vetores $(1, 1)$ e $(4, 1)$ formam uma base para o ${\mathbb R}^2$, e como
$
\begin{eqnarray}
T({\bf v}_1) &=& 1{\bf v}_1 + 0{\bf v}_2 \nonumber \\
T({\bf v}_2) &=& -2{\bf v}_2 = 0{\bf v}_1 - 2{\bf v}_2 \nonumber
\end{eqnarray}
$
temos $[T]_\beta^\beta = \left[\begin{array}{c c} 1 & 0 \\ 0 & -2\end{array}\right]$.

Podemos mostrar que dada uma transformação linear $T: V \rightarrow V$, se conseguirmos uma base $\beta =\{{\bf v}_1, ..., {\bf v}_n\}$ temos que
$$
\begin{eqnarray}
T({\bf v}_1) &=& \lambda_1{\bf v}_1 + 0{\bf v}_2 + ... + 0{\bf v}_n \nonumber \\
T({\bf v}_2) &=& 0 {\bf v}_1 + \lambda_ 2 {\bf v}_2 + ... + 0{\bf v}_n \nonumber \\
\vdots \\
T({\bf v}_n) &=& 0 {\bf v}_1 + 0{\bf v}_2 + ... + \lambda_n {\bf v}_n \nonumber 
\end{eqnarray}
$$

E a matriz $[T]_\beta^\beta$ será diagonal e dada por
$$
[T]_\beta^\beta = \left[\begin{array}{c c c c} \lambda_1 & 0 & ... & 0 \\ 0 & \lambda_2 & ... & 0 \\ 0 & 0 & ... & \lambda_n\end{array}\right]
$$

**Exemplo:** $A = \left[\begin{array}{c c c} 1 & 2 & 0 \\ 2 & 1 & 0 \\ 0 & 0 & 1\end{array}\right]$ tem polinômio característico $p(\lambda) = (\lambda -1)(\lambda + 1)(\lambda -3)$, e assim três autovalores distintos, logo 
$$[T]_\beta^\beta = \left[\begin{array}{c c c} 1& 0 &  0 \\ 0 & -1 & 0 \\ 0 & 0 & 3\end{array}\right]$$


**Exemplo:** $A = \left[\begin{array}{c c c} -1 & 2 & 2 \\ 2 & -1 & 2 \\ 2 & 2 & -1\end{array}\right]$ tem polinômio característico $p(\lambda) = (\lambda - 3)^2(\lambda + 3)$, e assim um autovalor com multiplicadade 2, e outro com multiplicidade 1. 

Determinar os autovetores associados
$
[T]_\beta^\beta = \left[\begin{array}{c c c} -3& 0 &  0 \\ 0 & -3 & 0 \\ 0 & 0 & 3\end{array}\right]
$


Observe que nesse último exemplo, que um autovalor aparece na diagonal tantas vezes quantas forem os autovalores LI.

**Definicao:** Seja $T: V \rightarrow V$ um operador linear, dizemos que $T$ é diagonalizável se existe uma base de $V$ cujos elementos são autovalores de $T$.


Observações:
1. Veja que isso é equivalente a dizer que se uma matriz tem $n$ autovalores distintos ele é diagonalizável.
2. Porém, o fato da mariz $n\times x$ ter menos que $n$ autovalores distintos não necessariamente implica que ela não é diagonalizável, um exemplo simples é a matriz identidade, que é diagonal, e só tem um autovalor.

**Exemplo:** As matrizes abaixo são diagonalizáveis? $A = \left[\begin{array}{c c c}1 & 3 & 0 \\ 0 & 2 & 1\\ 0 & 1 & 1 \end{array}\right]$, e $B = \left[\begin{array}{c c c}1 & 0 & 0 \\ 0 & -2 & 1\\ 1 & 0 & 1 \end{array}\right]$

Vamos determinar os autovalores de $A$: 
$
p(\lambda) = det(\lambda - A) = det\left[\begin{array}{c c c}\lambda - 1 & -3 & 0 \\ 0 &\lambda -  2 & -1\\ 0 & -1 & \lambda - 1 \end{array}\right]
$
e
$p(\lambda) = (\lambda -1) (\lambda - \frac{2-\sqrt{5}}{2})(\lambda + \frac{3+\sqrt{5}}{2})$, logo tem três autovalores, e é diagonalizável.

Para $B$ temos 
$
p(\lambda) = det(\lambda - B) = det\left[\begin{array}{c c c}\lambda - 1 & 0 & 0 \\ 2 &\lambda -  1 & 0\\ -1 & 0 & \lambda - 1 \end{array}\right]
$
$p(\lambda) = (\lambda -1)^3$, e $\lambda = 1$ é um autovalor com multiplicidade 3. Portanto, não podemos garantir que $B$ é diagonalizável, vamos determinar os autovalores:
$
\left[\begin{array}{c c c}1 & 0 & 0 \\ -2 & 1 & 0 \\1 & 0 & 1\end{array}\right]\left[\begin{array}{c} x \\ y \\ z\end{array}\right] = \left[\begin{array}{c} x \\ y \\ \\ z \end{array}\right]
$
logo temos
$\left[\begin{array}{c} 0 \\ y \\ y\end{array}\right] = y\left[\begin{array}{c} 0 \\ 1 \\ 1\end{array}\right]  + z\left[\begin{array}{c} 0 \\ 0 \\ 1\end{array}\right] $
e como temos dois autovalores, não temos uma base para o ${\mathbb R}^3$, assim $B$ não é diagonalizável.

## <font color="0000ff">Polinômio Minimal</font>

Vimos que se $Dim(V) = n$ e operador linear $T:V \rightarrow V$ tem $n$ autovalores distintos, então a matriz desse operador é diagonalizável. Vamos determinar o polinômio minimal do operador $T$.

**Definicao:** Seja $p(x) = a_nx^n + ... + a_1x + a_0$ e uma matriz quadrada, então $P(A)$ é a matriz
$
P(A) = a_nA^n + ... + a_1A + a_oI_{n\times n}
$
e quando $p(A) = 0$, dizemos que o polinômio anula a matriz $A$.



**Exemplo:**  $A = \left[\begin{array}{c c} -1 & 4  \\ 2 & 1 \end{array}\right]$, $p(x) = x^2 - 9$, $q(x) = 2x+3$ $p(x)$ anula $A$, $q(x)$ não anula $A$.

**Definicao:** Seja $A$ uma matriz quadrada, o polinômio minimal de $A$  é dado por
$m(x) = x^k + a_{k-1}x^{k-1} + ... + a_0$, tal que 
* $m(A) = 0$
* $m(x)$ é o polinômio minimal de menor grau entre aqueles que anulam $A$.

**Tteorema:** Seja $T: V \rightarrow V$ um operador linear, $\alpha$ uma base de $V$ e $p(\alpha)$ o polinômio característico de $T$, então $p([T])_{\alpha}^{\alpha}) = 0$.


**Teorema:** As raízes do polinômio minimal são as mesmas raízes (distintas) do polinômio característico.


Assim, dado o polinômio característico, o polinômio minimal terá grau igual ou menor que o polinômio característico. Para verificar qual dos polinômios com as mesmas raízes que o polinômio característico é o minimal, devemos testar qual dos que tem o menor grau que anula a matriz $T$.


**Teorema:** Seja $\lambda_1, ..., \lambda_n$ autovalores distintos de um operador linear $T$. Então $T$ será diagonalizável se, e somente se, o polinômio $(x - \lambda_1)(x - \lambda_2)...(x - \lambda_n)$ anula a matriz $T$.


**Exemplo:** Determine o polinômio minimal $m(x)$ de cada uma das matrizes $A = \left[\begin{array}{c c} \lambda & 0  \\ 0 & \lambda \end{array}\right]$.
