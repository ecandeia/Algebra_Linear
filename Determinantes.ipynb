# Determinantes

Dada uma matriz quadrada ${\bf A}_{N\times N}$ define-se o seu determinante como
$ det {\bf A}_{N\times N} = \sum_{j=1}^{N}a_{ij}\Delta_{ij}
$
sendo $\Delta_{ij} = (-1)^{i+j}|A_{ij}|$ o co-fator e $|A_{ij}|$ o determinante da submatriz obtida retirando a $i$-ésima linha e $j$-ésima coluna.

**Exemplo:**
$
A = \left[
\begin{array}{c c}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{array}
\right]
$

**Exemplo:**
$
A = \left[
\begin{array}{c c c}
3 & 7 & 9  \\
-1 & 2 & 6 \\
4 & 3 & 2
\end{array}
\right]
$

O desenvolvimento de Laplace é um procedimento recursivo.

## Propriedades:
* Se todos os elementos de uma linha (coluna) forem nulos, $det(A) = 0$.
* $det(A) = det(A^T)$.
* Se multiplicarmos uma linha de uma matriz por uma constante, o determinante fica multiplicado por essa constante.
 
**Definição:** [Matriz dos cofatores] $\overline{A} = [\Delta_ij]$.

**Definição:** [Matriz Adjunta] Para uma matriz quadrada $adj(A) = \overline{A}^T = [\Delta_ij]^T$.

**Exemplo:** 
${\bf A} = \left[
\begin{array}{c c c}
2 & 1 & 0 \\
-3 & 1 & 4 \\
1 & 6 & 5 \\
\end{array}
\right]
$

**Teorema:** $A\times\overline{A}^T = det(A)I_n$

## Matriz Inversa

**Definição:** A matriz inversa de ${\bf A}_{n\times n}$ é uma matriz ${\bf B} = {\bf A} ^{-1}$ tal que
$
%{\bf A}\cdot{\bf B} = {\bf B}\cdot{\bf A} = {\bf I}_n
$

**Exemplo:*
$
{\bf A} = \left[
\begin{array}{c c}
3 & 1\\
5 & -1
\end{array}
\right]
$

Observações:
* Qual número aparece no denominador dos exemplos anteriores?
* Nem toda matriz tem inversa. 
* Se {\bf A} e {\bf B} são quadradas de mesma dimensão, e ambas inversíveis, ent\~{ao}o
    * $({\bf A}\cdot {\bf B}) ^{-1} = {\bf B}^{-1}\cdot {\bf A}^{-1}$.
    * $({\bf A}\cdot{\bf B})({\bf B}^{-1}\cdot {\bf A}^{-1}) = {\bf A}({\bf B}\cdot{\bf B})^{-1}{\bf A}^{-1} = {\bf A}\cdot {\bf I}\cdot{\bf A}^{-1} = {\bf A}\cdot{\bf A}^{-1} = {\bf I}$
* Se ${\bf A}^{-1}$ existe: $det({\bf A}\cdot {\bf A}^{-1}) = det{\bf A}\cdot{\bf A}^{-1} = det({\bf I}_n) = 1$, logo $det({\bf A})\cdot det({\bf A}^{-1}) = 1$
    * $det({\bf A}) \neq 0$
    * $det({\bf A}^{-1})  = \frac{1}{det({\bf A})}$ e $det({\bf A})$ é uma condição necessária para que a inversa exista.

**Teorema:**
Uma matriz quadrada $\bf A$ admite inversa se e somente se, $det({\bf A}) \neq 0$ e 
$
{\bf A}^{-1} = \frac{1}{det{\bf A}}adj({\bf A})
$

## Determinantes - Regra de Cramer

Seja um sistema com $n$ equações e $n$ incógnitas.
$
\left\{
\begin{array}{c}
a_{11}x_1 + .... + a_{1n}x_n = b_1 \\
\vdots \\
a_{n1}x_1 + ... + a_{nn}x_n = b_n
\end{array}
\right.
$
escrevendo na forma matricial
$
\left[
\begin{array}{c}
a_{11} + .... + a_{1n} \\
\vdots \\
a_{n1} + ... + a_{nn}
\end{array}\right]
\left[
\begin{array}{c}
x_1 \\ \vdots \\ x_n
\end{array}\right]
= 
\left[\begin{array}{c}
b_1 \\ \vdots \\ b_n
\end{array}\right]
$
ou $Ax = b$

Vamos assumir que $det(A) \neq 0$, logo $A^{-1}$ existe e
$
\begin{eqnarray}
A^{-1}(Ax) = A^{-1}b \nonumber \\
(A^{-1}A)x = A^{-1}b \nonumber \\
I_n x = A^{-1}b \nonumber \\
x = A^{-1}b  \nonumber
\end{eqnarray}
